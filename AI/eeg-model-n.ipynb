## üß† EEG Signal Classes ‚Äì Brief Description

- **Seizure**  
  Seizures are abnormal bursts of electrical activity in the brain. They often indicate epilepsy or acute neurological events. Recognizing seizure patterns is critical for immediate diagnosis and treatment.

- **LPD (Lateralized Periodic Discharges)**  
  LPDs are repetitive, periodic waveforms occurring on one side of the brain. They may be associated with acute brain injuries, such as strokes, infections, or tumors.

- **GPD (Generalized Periodic Discharges)**  
  GPDs are periodic discharges that occur across both hemispheres. They are typically linked with diffuse brain dysfunction and may appear in metabolic encephalopathy or post-anoxic brain injury.

- **LRDA (Lateralized Rhythmic Delta Activity)**  
  LRDA refers to rhythmic slow wave activity (delta range) occurring on one hemisphere. It may reflect structural brain abnormalities or evolving seizure activity.

- **GRDA (Generalized Rhythmic Delta Activity)**  
  GRDA involves rhythmic delta activity distributed symmetrically across the brain. It can be seen in encephalopathy or during the recovery phase after a seizure.

- **Other**  
  This class includes EEG segments that do not fit into the defined categories or represent normal/ambiguous activity. It helps reduce false positives and improves model robustness.


## üß† What is the "Double Banana" Montage in EEG?

The **"double banana"** is a common EEG montage (electrode arrangement) used to visualize brain activity. It is a **longitudinal bipolar montage**, meaning each channel represents the voltage difference between two electrodes placed **in a straight line along the anterior-posterior axis** of the head.

- The term ‚Äúbanana‚Äù comes from the arc-shaped pattern of the electrodes on each hemisphere.
- ‚ÄúDouble‚Äù refers to having one arc on the left hemisphere and another on the right, forming two curved lines ‚Äî hence "double banana".

### üìå Key Features:
- Helps in identifying lateralized abnormalities (e.g., seizures or slowing on one side of the brain).
- Makes it easier to spot phase reversals, which indicate the location of maximum voltage change (useful in localizing seizure onset).
- Standard montage used in clinical EEG recordings and educational materials.

## üìä Spectrograms in EEG Analysis

A **spectrogram** is a visual representation of the **frequency content of a signal over time**. In EEG analysis, spectrograms are widely used to transform raw brainwave signals into 2D images where:

- The **x-axis** represents time  
- The **y-axis** represents frequency (in Hz)  
- The **color intensity** represents power (amplitude) at a specific frequency and time

---

### üß† Why Use Spectrograms for EEG?

EEG signals are **non-stationary**, meaning their frequency content changes over time. By converting the signals into spectrograms, we can:

- Capture temporal and spectral information simultaneously
- Detect frequency-based patterns such as seizures, slowing, or rhythmic discharges
- Enable the use of image-based deep learning models like CNNs

---

### üß© Brain Region Labels

Each spectrogram corresponds to a recording from a specific **brain region or electrode group**, often labeled as:

| Label | Region Description                     |
|-------|----------------------------------------|
| **LL** | Left Lateral ‚Äì Left side, outer region of the brain |
| **RL** | Right Lateral ‚Äì Right side, outer region of the brain |
| **LP** | Left Parasagittal ‚Äì Left side, near the midline |
| **RP** | Right Parasagittal ‚Äì Right side, near the midline |

These spatial labels help the model learn **location-based patterns** that might be indicative of specific neurological conditions.

## üßÆ Step 2 ‚Äì Calculate Start and End Rows for Spectrogram Slices

Each sub-spectrogram in the dataset is 600 seconds long, and each row in the full spectrogram file represents 2 seconds.  
So, we divide the offset (in seconds) by 2 to get the `start_row`, and since each segment spans 600 seconds (which is 300 rows), we add 300 to get the `end_row`.

```python
# Calculate the starting row in the full spectrogram file
# Each 2 seconds = 1 row ‚áí divide seconds by 2
df["start_row"] = df["spectrogram_label_offset_seconds"] / 2

# Each sub-spectrogram spans 600 seconds ‚áí 600 / 2 = 300 rows
# So the end row is start_row + 300
df["end_row"] = df["start_row"] + 300

# Display the updated DataFrame to confirm changes
df.head()
```

## ‚úÇÔ∏è Step 3 ‚Äì Use a Subset of the Dataset for Faster Processing

For faster experimentation or limited resource environments (like Kaggle kernels), we may choose to work on a smaller subset of the full dataset.

```python
# Select the first 25,082 rows from the training data
# This helps reduce processing time and memory usage during development
half_data = df.head(25082)

# Reset the index of the subset 
half_data = half_data.reset_index(drop=True)

# Display the first few rows of the subset
half_data.head()
```

## üß™ Step 4 ‚Äì Convert Spectrograms to NumPy Arrays (`.npy`)

This loop processes each EEG segment by:
- Loading the corresponding `.parquet` spectrogram file
- Extracting the correct time window using `start_row` and `end_row`
- Handling missing values
- Saving the result as a `.npy` file for efficient model input

This step speeds up training and avoids redundant computation during each epoch.

```python
# Required libraries
import numpy as np
import os
from tqdm import tqdm

# === Step 4: Processing EEG Spectrograms ===
processed_df_list = []  # To store paths and labels for processed files

# Iterate through each row of the dataset with a progress bar
for _, row in tqdm(half_data.iterrows(), total=half_data.shape[0], desc="Processing Spectrograms"):
    
    # Extract spectrogram and label IDs
    spec_id = row['spectrogram_id']
    label_id = row['label_id']
    
    # Build file paths
    input_spec_path = base_path / "train_spectrograms" / f"{spec_id}.parquet"
    output_npy_path = output_path / f"{label_id}.npy"

    # Skip if already processed
    if not os.path.exists(output_npy_path):
        try:
            # Load the .parquet spectrogram file
            spectrogram = pd.read_parquet(input_spec_path)

            # Extract the slice corresponding to the sub-segment (600 seconds ‚Üí 300 rows)
            start = int(row["start_row"])
            end = int(row["end_row"])
            spectrogram = spectrogram.iloc[start:end]

            # Fill any NaN values with column-wise means (prevents model failure)
            spectrogram = spectrogram.fillna(spectrogram.mean())

            # Drop the 'time' column and convert to NumPy array
            npy_data = spectrogram.drop('time', axis=1).to_numpy()

            # Save the array as .npy for faster loading during training
            np.save(output_npy_path, npy_data)

        except FileNotFoundError:
            print(f"Missing file for spectrogram_id: {spec_id}. Skipping.")
            continue

    # Append output path and label to list for later use
    processed_df_list.append({
        'npy_path': str(output_npy_path),
        'expert_consensus': row['expert_consensus']
    })

# Convert the list of dictionaries into a DataFrame for downstream use
processed_df = pd.DataFrame(processed_df_list)
print(f"\nCreated {len(processed_df)} .npy files.")
processed_df.head()
```

## üì¶ Step 9 ‚Äì Create Custom Dataset Class for EEG Spectrograms

This PyTorch `Dataset` class handles loading and preprocessing of spectrogram data.  
It ensures that each sample has a fixed shape, applies optional augmentation, and returns data-label pairs.

```python
import torchvision.transforms as T
from torch.utils.data import Dataset, DataLoader

class SpectrogramDataset(Dataset):
    def __init__(self, df, augment=False):
        self.df = df
        self.augment = augment

        # Define simple torchvision-style augmentations (applied only if augment=True)
        self.transforms = T.Compose([
            T.RandomHorizontalFlip(p=0.5),
            T.RandomVerticalFlip(p=0.5),
            T.RandomAffine(degrees=10, translate=(0.1, 0.1), scale=(0.9, 1.1)),
        ])

    def __len__(self):
        # Return total number of samples
        return len(self.df)

    def __getitem__(self, idx):
        # Load .npy spectrogram file for the given index
        path = self.df.iloc[idx]["npy_path"]
        data = np.load(path)

        # Normalize: zero mean, unit variance
        data = (data - data.mean()) / (data.std() + 1e-8)

        # Ensure fixed shape: width = 400 columns (time), height = 300 rows (frequencies)
        if data.shape[1] < 400:
            pad_cols = 400 - data.shape[1]
            data = np.pad(data, ((0, 0), (0, pad_cols)), mode='constant')
        elif data.shape[1] > 400:
            data = data[:, :400]

        if data.shape[0] < 300:
            pad_rows = 300 - data.shape[0]
            data = np.pad(data, ((0, pad_rows), (0, 0)), mode='constant')
        elif data.shape[0] > 300:
            data = data[:300, :]

        # Reshape into 3D format: [channels, height, width]
        # Here: [4, 300, 100] by slicing into 4 equal vertical segments (simulating multi-channel input)
        try:
            data = data.reshape(300, 4, 100).transpose(1, 0, 2)  # shape: [4, 300, 100]
        except Exception as e:
            raise ValueError(f"‚ùå Error reshaping file {path} with shape {data.shape}: {e}")

        # Convert to PyTorch tensor
        data = torch.from_numpy(data).float()

        # Apply augmentation if enabled
        if self.augment:
            data = self.transforms(data)

        # Get label and convert to tensor
        label = torch.tensor(self.df.iloc[idx]["class"], dtype=torch.long)

        return data, label
```

## üß† Step 11 ‚Äì Define CNN Model for EEG Spectrogram Classification

This Convolutional Neural Network (CNN) is designed to classify EEG spectrogram inputs into 6 medical categories.  
It takes an input of shape `(Batch, 4, 300, 100)` ‚Äî representing 4 pseudo-channels of spectrogram data ‚Äî  
and passes it through several convolutional blocks followed by fully connected layers.

```python
import torch
import torch.nn as nn

class EEG_Spectrogram_CNN(nn.Module):
    def __init__(self, num_classes=6):
        super(EEG_Spectrogram_CNN, self).__init__()

        # === Convolutional Block 1 ===
        # Input: (Batch, 4, 300, 100)
        self.conv1 = nn.Sequential(
            nn.Conv2d(in_channels=4, out_channels=32, kernel_size=(5, 5), stride=1, padding=2),  # ‚Üí (Batch, 32, 300, 100)
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=(2, 2), stride=2)  # ‚Üí (Batch, 32, 150, 50)
        )

        # === Convolutional Block 2 ===
        self.conv2 = nn.Sequential(
            nn.Conv2d(in_channels=32, out_channels=64, kernel_size=(3, 3), stride=1, padding=1),  # ‚Üí (Batch, 64, 150, 50)
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=(2, 2), stride=2)  # ‚Üí (Batch, 64, 75, 25)
        )

        # === Convolutional Block 3 ===
        self.conv3 = nn.Sequential(
            nn.Conv2d(in_channels=64, out_channels=128, kernel_size=(3, 3), stride=1, padding=1),  # ‚Üí (Batch, 128, 75, 25)
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=(3, 3), stride=3)  # ‚Üí (Batch, 128, 25, 8)
        )

        # Flatten layer to prepare for fully connected classification head
        self.flatten = nn.Flatten()  # Output size: 128 √ó 25 √ó 8 = 25,600 features

        # === Fully Connected Classifier ===
        self.classifier = nn.Sequential(
            nn.Linear(128 * 25 * 8, 1024),
            nn.ReLU(),
            nn.Dropout(0.5),  # Dropout for regularization
            nn.Linear(1024, num_classes)  # Output: logits for 6 classes
        )

    def forward(self, x):
        # Forward pass through the network
        x = self.conv1(x)  # ‚Üí (Batch, 32, 150, 50)
        x = self.conv2(x)  # ‚Üí (Batch, 64, 75, 25)
        x = self.conv3(x)  # ‚Üí (Batch, 128, 25, 8)
        x = self.flatten(x)
        logits = self.classifier(x)
        return logits
```


Let's break down the model's workflow step-by-step, from input to output.

### High-Level Overview

The model takes a 4-channel array as input and processes it through several stages to produce 6 output values. This structure is typical for a classification problem with 6 possible categories. The architecture consists of two main parts:
1.  **Feature Extraction:** A series of Convolutional, ReLU, and Max Pooling layers that automatically learn and identify features (like edges, textures, and shapes) from the input image.
2.  **Classification:** A set of fully connected layers that take the extracted features and perform the final classification.

---

### Detailed Layer-by-Layer Explanation

1.  **Input (input.1)**
    *   **Shape:** `1 √ó 4 √ó 300 √ó 100`
    *   **What it is:** This is the input to the network. The dimensions represent:
        *   `1`: Batch size (processing one array at a time) , it can be 16,32,64 or any number of batches.
        *   `4`: Number of channels: LL = left lateral; RL = right lateral; LP = left parasagittal; RP = right parasagittal.
        *   `300`: Height of the array.
        *   `100`: Width of the array.

2.  **First Convolutional Block**
    *   **Conv (Convolutional Layer):** This layer applies 32 different filters (or kernels) of size `5√ó5` to the input image. Each filter is designed to detect a specific low-level feature (e.g., a horizontal edge, a specific color gradient).
        *   `W <32√ó4√ó5√ó5>`: These are the filter weights. 32 filters, each processing 4 input channels, with a spatial size of 5x5.
        *   `B <32>`: A bias term is added to the output of each of the 32 filters.
    *   **Relu (Rectified Linear Unit):** This is an activation function. It introduces non-linearity by changing all negative values in the feature maps to zero. This helps the model learn more complex patterns.
    *   **MaxPool (Max Pooling):** This layer downsamples the feature maps, reducing their height and width. It does this by taking the maximum value from a small window as it slides across the map. This makes the model more efficient and robust to small variations in the position of features.

3.  **Second Convolutional Block**
    *   **Conv:** The model now applies a deeper set of 64 filters, each with a `3√ó3` size, to the feature maps from the previous block. Smaller filters in deeper layers tend to combine the simpler features into more complex ones.
        *   `W <64√ó32√ó3√ó3>`: 64 filters, processing the 32 feature maps from the previous block.
        *   `B <64>`: 64 bias terms.
    *   **Relu & MaxPool:** These perform the same functions as before, further processing and downsampling the 64 new feature maps.

4.  **Third Convolutional Block**
    *   **Conv:** The complexity increases again. This layer uses 128 filters of size `3√ó3`.
        *   `W <128√ó64√ó3√ó3>`: 128 filters, processing the 64 feature maps from the previous block.
        *   `B <128>`: 128 bias terms.
    *   **Relu & MaxPool:** The final activation and downsampling step in the feature extraction part. At this point, the model has produced 128 highly abstract feature maps of a much smaller spatial size than the original image.

5.  **Flatten Layer**
    *   **What it does:** This layer takes the final 3D block of feature maps (e.g., shape `128 √ó height √ó width`) and unrolls it into a single, long 1D vector. This is necessary to transition from the convolutional part of the network to the fully connected classification part. The next layer's input size (`25600`) reveals the size of this flattened vector.

6.  **First Fully Connected Block**
    *   **Gemm (General Matrix Multiplication):** This is a fully connected (or dense) layer. It performs a matrix multiplication to transform the 25,600-element feature vector into a 1,024-element vector. This layer learns to weigh the importance of different features for the classification task.
        *   `B <1024√ó25600>`: The weight matrix. 
        *   `C <1024>`: The bias vector.
    *   **Relu:** Another activation function is applied to the 1,024-element vector.

7.  **Second Fully Connected Block (Output Layer)**
    *   **Gemm:** This is the final layer of the network. It takes the 1,024-element vector and transforms it into the final output vector of size 6.
        *   `B <6√ó1024>`: The weight matrix that maps the 1,024 features to 6 output classes.
        *   `C <6>`: The final bias vector for the 6 classes.

8.  **Output (Node 41)**
    *   **Shape:** `1 √ó 6`
    *   **What it is:** The model's final output is a vector of 6 numbers. In a classification context, each number represents a "score" or "logit" for one of the 6 possible classes. To get probabilities, these scores are typically passed through a Softmax function (which is not shown in this diagram). The class with the highest score (or probability) is the model's prediction.

### Summary of the Process

The network progressively transforms a large, high-dimensional image into a small, 6-element vector. It starts by finding simple patterns, combines them into more complex features, and finally uses those features to make a decision about which of the 6 categories the original image belongs to.

- **Validation** is done with `torch.no_grad()` to disable gradient tracking

Intermediate progress is printed using `tqdm` and occasional batch updates.

```python
from tqdm import tqdm

epochs = 12

for epoch in tqdm(range(epochs)):
    # === Training Phase ===
    model.train()
    train_loss = 0.0
    train_accuracy = 0.0

    for batch, (x, y) in enumerate(train_loader):
        x = x.to(device)
        y = y.to(device)

        # Zero the parameter gradients
        optimizer.zero_grad()

        # Forward pass
        y_pred = model(x)

        # Compute loss
        loss = loss_fn(y_pred, y)

        # Backward pass (compute gradients)
        loss.backward()

        # Update model parameters
        optimizer.step()

        # Accumulate loss and accuracy
        train_loss += loss.item()
        correct = (y_pred.argmax(dim=1) == y).sum().item()
        train_accuracy += correct / y.size(0)

        # Optional progress print every ~5000 samples
        if batch % 157 == 0:
            print(f"{batch * BATCH_SIZE} from {len(train_loader) * BATCH_SIZE}")

    # Average metrics over all batches
    train_loss /= len(train_loader)
    train_accuracy /= len(train_loader)

    print(f"[Epoch {epoch+1}/{epochs}] Train Loss: {train_loss:.4f}, Train Accuracy: {train_accuracy:.4f}")

    # === Validation Phase ===
    model.eval()
    val_loss = 0.0
    val_accuracy = 0.0

    with torch.no_grad():  # Disable gradient tracking for validation
        for batch, (x, y) in enumerate(val_loader):
            x = x.to(device)
            y = y.to(device)

            # Forward pass
            y_pred = model(x)
            loss = loss_fn(y_pred, y)

            # Accumulate loss and accuracy
            val_loss += loss.item()
            correct = (y_pred.argmax(dim=1) == y).sum().item()
            val_accuracy += correct / y.size(0)

            # Optional progress print
            if batch % 25 == 0:
                print(f"{batch * BATCH_SIZE} from {len(val_loader) * BATCH_SIZE}")

    # Average validation metrics
    val_loss /= len(val_loader)
    val_accuracy /= len(val_loader)

    print(f"         >> Validation Loss: {val_loss:.4f}, Validation Accuracy: {val_accuracy:.4f}")
```

## üìä Step 15 ‚Äì Confusion Matrix Visualization

After training, it's important to assess the model‚Äôs **per-class performance**.  
The **confusion matrix** shows how often each class was correctly predicted or confused with another class.

In this case:
- We normalize by row (`normalize='true'`) to see **percentages relative to true labels**
- The matrix uses the original class names (e.g., "GPD", "Seizure") from `LabelEncoder`

This helps identify which classes are most frequently misclassified.

```python
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# === Step 1: Collect true and predicted labels ===
# Make sure your model is in evaluation mode and not tracking gradients
model.eval()
all_preds = []
all_labels = []

with torch.no_grad():
    for x, y in val_loader:
        x = x.to(device)
        y = y.to(device)

        outputs = model(x)
        preds = torch.argmax(outputs, dim=1)

        # Store predictions and true labels
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(y.cpu().numpy())

# === Step 2: Plot Normalized Confusion Matrix (as % per true class) ===
ConfusionMatrixDisplay.from_predictions(
    y_true=all_labels,
    y_pred=all_preds,
    normalize='true',              # Normalize by true labels (row-wise)
    values_format='.0%',           # Display values as whole number percentages
    display_labels=le.classes_     # Use the original class names from LabelEncoder
)

# === Step 3: Plot Formatting ===
plt.title("Normalized Confusion Matrix (%)")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## ‚ùå Step 16 ‚Äì Confusion Matrix for Misclassifications Only

To focus purely on the **model's mistakes**, we compute a confusion matrix where:

- Each correctly predicted sample has weight 0  
- Each incorrect sample has weight 1 (`sample_weight`)  

This reveals **only where and how often the model makes errors**, helping us identify problematic class pairs.

```python
all_labels = np.array(all_labels)
all_preds = np.array(all_preds)

# Highlight errors only
sample_weight = (all_labels != all_preds).astype(int)

# Plot confusion matrix using errors only
ConfusionMatrixDisplay.from_predictions(
    y_true=all_labels,
    y_pred=all_preds,
    sample_weight=sample_weight,
    normalize='true',
    values_format='.0%',
    display_labels=le.classes_
)

plt.title("Confusion Matrix (Misclassifications Only, %)")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## üß™ Step 17 ‚Äì Inference on a Single Test Spectrogram

We now load a `.parquet` spectrogram file from the test set, preprocess it exactly as during training, and pass it through the trained model to get a prediction.

Steps:
1. Load and normalize the EEG spectrogram
2. Reshape into CNN-compatible format: `(4, 300, 100)`
3. Convert to PyTorch tensor and add batch dimension
4. Get predicted class from model output

```python
# === Load and preprocess test sample ===
test_df = pd.read_parquet("/kaggle/input/hms-harmful-brain-activity-classification/test_spectrograms/853520.parquet")

# Drop the 'time' column and convert to NumPy array
test_array = np.array(test_df.iloc[:, 1:])

# Normalize (zero mean, unit variance)
test_array = (test_array - test_array.mean()) / (test_array.std() + 1e-8)

# Reshape to match model input: (4, 300, 100)
test_array = test_array.reshape(300, 4, 100).transpose(1, 0, 2)

# Convert to torch tensor and move to device
test_tensor = torch.tensor(test_array, dtype=torch.float).to(device)

# Add batch dimension: shape becomes (1, 4, 300, 100)
test_tensor = test_tensor.unsqueeze(0)

# === Inference ===
model.eval()  # Ensure model is in evaluation mode
y_test_logit = model(test_tensor)
y_test_probs = torch.softmax(y_test_logit, dim=1)
predicted_class = torch.argmax(y_test_probs, dim=1)

# Display the predicted class index
predicted_class
```

**`853520.parquet`** from the test set.

üìå **Prediction Result:**
- üî¢ Predicted Class Index: `4`
- üè∑Ô∏è Predicted Label: **Other**
- ‚úÖ This prediction is **correct** based on the ground truth.

This demonstrates the model's ability to generalize to unseen EEG segments and correctly identify non-pathological or undefined brain activity labeled as **"Other"**.
